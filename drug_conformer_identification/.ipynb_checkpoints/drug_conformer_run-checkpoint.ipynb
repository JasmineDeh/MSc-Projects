```python
# This cell imports all of the modules and functions needed to complete this project.
import numpy as np
import pandas as pd
import big_project_functions as pf
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Data read in using pandas.read_csv.
drug_absent = pd.read_csv('no_drug.csv')
drug_present = pd.read_csv('drug.csv')
```

```python
# .head() used to view head of data to see if it imported correctly.
drug_absent.head()
```

```python
drug_present.head()
```

```python
# pf.pull_array function pulls x, y, z pandas.Series from main .csv file and returns numpy array.
# pf.distances function calculates the distances of 2 arrays, returns a pandas.DataFrame of the results.
# Data column is then renamed to match amino acid pairs.
gly1abs_arg1abs = pf.distances((pf.pull_array(drug_absent, 'Gly1')), (pf.pull_array(drug_absent, 'Arg1')))

gly1a_arg1a = gly1abs_arg1abs.rename(columns={0: 'Gly1a_Arg1a'})
```

```python
gly2abs_ala1abs = pf.distances((pf.pull_array(drug_absent, 'Gly2')), (pf.pull_array(drug_absent, 'Ala1')))

gly2a_ala1a = gly2abs_ala1abs.rename(columns={0: 'Gly2a_Ala1a'})
```

```python
thr1abs_his1abs = pf.distances((pf.pull_array(drug_absent, 'Thr1')), (pf.pull_array(drug_absent, 'His1')))

thr1a_his1a = thr1abs_his1abs.rename(columns={0: 'Thr1a_His1a'})
```

```python
ala2abs_pro1abs = pf.distances((pf.pull_array(drug_absent, 'Ala2')), (pf.pull_array(drug_absent, 'Pro1')))

ala2a_pro1a = ala2abs_pro1abs.rename(columns={0: 'Ala2a_Pro1a'})
```

```python
# pandas.concat function used to join distance DataFrames together, as they have no common column to merge on.
# 'axis=1' so that they join column by column.
drug_absent_dist = pd.concat([gly1a_arg1a, gly2a_ala1a, thr1a_his1a, ala2a_pro1a], axis=1)
drug_absent_dist
```

```python
# Histograms plotted for each pair to see distrbution and check for outliers.
# Gly1 Arg1 (absent) displays normal distribution.
fig, ax = plt.subplots(2, 2, figsize=(9,9))
ax[0][0].hist(drug_absent_dist['Gly1a_Arg1a'], bins=12)
ax[0][0].set(xlabel='Distance (1E-10m)', ylabel='Frequency', title='Gly1 Arg1 Distances (absent)')

# Gly2 Ala1 (absent) shows non-symmetric bimodal distribution, there are two clear seperate modes.
ax[0][1].hist(drug_absent_dist['Gly2a_Ala1a'], bins=12)
ax[0][1].set(xlabel='Distance (1E-10m)', ylabel='Frequency', title='Gly2 Ala1 Distances (absent)')

# Thr1 His1 (absent) displays almost normal distribution, the bell curve is not as strong as Gly1 Arg1 (absent).
ax[1][0].hist(drug_absent_dist['Thr1a_His1a'], bins=12)
ax[1][0].set(xlabel='Distance (1E-10m)', ylabel='Frequency', title='Thr1 His1 Distances (absent)')

# Ala2 Pro1 (absent) shows normal distibution, like Gly1 Arg 1 (absent).
ax[1][1].hist(drug_absent_dist['Ala2a_Pro1a'], bins=12)
ax[1][1].set(xlabel='Distance (1E-10m)', ylabel='Frequency', title='Ala2 Pro1 Distances (absent)')
```

```python
# pf.avg_stddv function to calculate means and standard deviations.
# The means are grouped around two main distances 6 and 8, suggesting potential for two clusters of data.
# Distributions in histograms explained by standard deviation scores, low score = normal distibution, higher score = broader distribution.
pf.avg_stddv(gly1a_arg1a, gly2a_ala1a, thr1a_his1a, ala2a_pro1a)
```

```python
# PCA conducted to further identify clusters and reduce dimensionality.
absent_pca = PCA(2)
absent_pca.fit(np.array(drug_absent_dist))
```

```python
absent_pca.components_
```

```python
# Two components make up over 85% of variance.
absent_pca.explained_variance_ratio_
```

```python
# PCA components plotted on scatter graph to see whether there are distict clusters.
x = absent_pca.fit_transform(drug_absent_dist)


fig, ax = plt.subplots()
ax.scatter(x[:, 0], x[:, 1])
ax.set(xlabel='Component 1', ylabel='Component 2', title='PCA of Distances (absent)')
plt.show()

# As expected two distict clusters were found.
# This was expected as the protein being explored is known to have two conformers in absence of the drug (normal conditions).
```

```python
# K-means clustering is conducted, using the expectation-maximisation (E-M) approach, to further define the clusters.

# Number of clusters are set and their centres are randomly intialised.
n_clusters = 2
i = np.random.randint(0, x.shape[0], size=n_clusters)
centres = x[i]

# While loop to perform E-M clustering method.
# E-step: each point is assigned to nearest cluster using 'vectors', 'magnitudes', and 'labels' variables.
# M-step: cluster centres are updated to mean of assigned points using 'new_centres' variable.
# np.all and break functions used to check if centres have stopped changing.
# 'centres' updated to 'new_centres' for the next iteration of the loop.
while True:
    vectors = x[:, np.newaxis] - centres
    magnitudes = np.linalg.norm(vectors, axis=-1)
    labels = magnitudes.argmin(axis=-1)

    new_centres = np.array([x[labels==i].mean(axis=0) for i in range(n_clusters)])

    if np.all(centres == new_centres):
        break

    centres = new_centres
```

```python
# Same procedure that was used for the 'drug_absent_dist' data, with updated variable names.
n_clusters_ = 3
j = np.random.randint(0, X.shape[0], size=n_clusters_)
centres_ = X[j]

while True:
    vectors_ = X[:, np.newaxis] - centres_
    magnitudes_ = np.linalg.norm(vectors_, axis=-1)
    labels_ = magnitudes_.argmin(axis=-1)

    new_centres_ = np.array([X[labels_==j].mean(axis=0) for j in range(n_clusters_)])

    if np.all(centres_ == new_centres_):
        break

    centres_ = new_centres_
```

```python
# Same procedure that was used for the 'drug_absent_dist' data.
fig, ax = plt.subplots(figsize=(6, 4))
for j in range(n_clusters_):
    ax.plot(X[labels_ == j, 0], X[labels_ == j, 1], '.', label=f'Conformer {j}')
ax.plot(centres_[:, 0], centres_[:, 1], 's', label='Final Centres')
plt.legend()
plt.show()

# Three clusters clearly defined as Conformer 0, Conformer 1, and Conformer 2.
```
