The prey_predator_populations.py model was created to hold the sparrows_hawks_cats($t$, $y$) function, $t$ = time value, and for this function specifically $y$ = the three population parameters. The new sparrows_hawks_cats($t$, $y$) function was created to later be used in the existing solve_ivp function. The new function sets the rules for the changes in the $y$ parameters in each time step. Therefore, by slicing the $y$ parameters so that sparrow_population is the first slice $y$[0], hawk_population is the second slice $y$[1], and cat_population is the third slice $y$[2], followed by the calculations for the different populations' birth and death rates, the new function can then return the changes in the three populations.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.integrate import solve_ivp
from prey_predator_populations import sparrows_hawks_cats
```

In the coded cell above, the models and functions needed to successfully run the new function have been imported.

```python
time_span = [0, 3000]
initial_populations = [100, 10, 0]

sol_populations = solve_ivp(sparrows_hawks_cats, time_span, initial_populations)

days_output = sol_populations.t
sparrow_change_output = sol_populations.y[0]
hawk_change_output = sol_populations.y[1]
cat_change_output = sol_populations.y[2]

print(days_output)
print(sparrow_change_output)
print(hawk_change_output)
print(cat_change_output)
```

In the coded cell above, the time step has been set to run for 3000 days. The initial_populations variable is the starting populations for the animals. The solve_ivp function allows for the change in populations to be explored, based on the time_span variable, initial_populations variable, and the parameters stated in the new sparrows_hawks_cats function. The output of the solve_ivp function is then split into different variables to individually view the days, change in sparrow population, change in hawk population, and change in cat population. These individual variables were then printed to check the functions were running correctly (as no change was expected it was easy to see if the output was correct).

```python
fig, ax = plt.subplots()

ax.plot(days_output, sparrow_change_output, label="Sparrow Population")
ax.plot(days_output, hawk_change_output, label="Hawk Population")
ax.plot(days_output, cat_change_output, label="Cat Population")

ax.set_xlabel('Days')
ax.set_ylabel('Number of Animals')
ax.set_title('Test 1')
ax.legend()

plt.show()
```

In the coded cell above, the previous days and populations' change output variables have been displayed in a graph (Test 1), to have better visual understanding of the data. In the Test 1 graph it can easily be seen that the system stayed stable as no change occured in the animal populations, based on the initial populations. This was expected as there were no cats to eat the sparrows, meaning that the sparrows were able to survive and repopulate at a rate that could supply the hawks with food. Therefore, continuing the harmonical balance between the hawk and sparrow populations.

```python
initial_populations_2 = [111, 19, 1]

sol_populations_2 = solve_ivp(sparrows_hawks_cats, time_span, initial_populations_2)

days_output_2 = sol_populations_2.t
sparrow_change_output_2 = sol_populations_2.y[0]
hawk_change_output_2 = sol_populations_2.y[1]
cat_change_output_2 = sol_populations_2.y[2]
```

In the coded cell above, the solve_ivp function has been ran again using the new sparrows_hawks_cats function, with the same time step as before, but with different initial populations. The same procedure that was previously used was carried out and the solve_ivp output was split so that the days and populations' outputs can be individually viewed in the graph below (Test 2).

```python
fig, ax = plt.subplots()

ax.plot(days_output_2, sparrow_change_output_2, label="Sparrow Population")
ax.plot(days_output_2, hawk_change_output_2, label="Hawk Population")
ax.plot(days_output_2, cat_change_output_2, label="Cat Population")

ax.set_xlabel('Days')
ax.set_ylabel('Number of Animals')
ax.set_title('Test 2')
ax.legend()

plt.show()
```

The above graph (Test 2) shows the roughly periodic system of the sparrow and hawk populations, as the populations grow and shrink over the 3000 day period. This was expected as the initial cat population was set to 1, because the cat feeds on the sparrows, it displaced the harmonical balance between the sparrow and hawk populations. The cat population does not change as it is not based on the sparrow or hawk populations. Even though the cat feeds on the sparrows, it does not rely on the sparrows as its only source of food.

```python
initial_populations_3 = [111, 19, 10]

sol_populations_3 = solve_ivp(sparrows_hawks_cats, time_span, initial_populations_3)

days_output_3 = sol_populations_3.t
sparrow_change_output_3 = sol_populations_3.y[0]
hawk_change_output_3 = sol_populations_3.y[1]
cat_change_output_3 = sol_populations_3.y[2]
```

Lastly, the coded cell above shows that the solve_ivp function was run for a third time, again using the new sparrows_hawks_cats function and the same time step previously used, but with different initial populations to the previous two times. Again, the same procedure was followed and the sliced solve_ivp outputs have been displayed in the graph below (Test 3).

```python
fig, ax = plt.subplots()

ax.plot(days_output_3, sparrow_change_output_3, label="Sparrow Population")
ax.plot(days_output_3, hawk_change_output_3, label="Hawk Population")
ax.plot(days_output_3, cat_change_output_3, label="Cat Population")

ax.set_xlabel('Days')
ax.set_ylabel('Number of Animals')
ax.set_title('Test 3')
ax.legend()

plt.show()
```

In the final graph (Test 3), a roughly decaying system is shown for the sparrows and hawks, as both populations drastically drop over the 3000 day period. This was expected as the initial cat population for this test was set to 10. The sparrows could not reproduce at a rate to supply both the cats and hawks with food, this then meant that the number of hawks dying off due to inadequate food supply grew faster than the reproduction rate of the hawks. Therefore, completely throwing off the natural balance between the hawk and sparrow populations. As previously stated the cat population does not change as they can acquire food from elsewhere.
